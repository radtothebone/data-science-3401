**Question 2.** Write a function named average that takes a list of numbers and returns their average. (10 Points)

```python
# The average function. 
def average(*args):
    return sum(args)/len(args)

# Running the function. 
average(5,50)
```

**Question 3.**  Convert a given number of days into a combination of weeks and days. (10 Points)

```python
# The conversion function. 
def days_to_weeks_and_days(days):
    # We will use floor division and the mod operator. 
    week_count = days // 7
    leftover_days = days % 7
    # Returning a tuple so we can separately address the values. 
    return week_count, leftover_days

# Making the number of days to convert a variable. 
days_to_convert = 24

# The tuple we will reference in the string. 
days_tuple = days_to_weeks_and_days(days_to_convert)

print(f'{days_to_convert} days is equal to {days_tuple[0]} weeks and {days_tuple[1]} days.')
```

**Question 4.** Write a function named string_length that takes a string and returns its length without using the built-in len() function. (10 Points)

```python
# The string function. 
def string_length(input_string):
    # Initializing the variable. 
    string_count = 0
    # Iterating over the string. 
    for i in input_string:
        string_count += 1
    # Returning the string count. 
    return string_count

string_length('Hello!')
```
