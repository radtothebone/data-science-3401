22. Write a Python script that asks the user for their current age, and then prints what their age will be in 10 years. (5 Points)

```python
user_age = input('Enter your age: ') # User age input.
age_in_10_years = int(user_age) + 10 # Age in 10 years calculation.
if age_in_10_years > 90:
    print('Wow, your age in 10 years is to high to calculate.') # Output if age exceeds 90.
else:
    print(f'In 10 years, you will be {age_in_10_years} years old.') # Output the future age.
```
