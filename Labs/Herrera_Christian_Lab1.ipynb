**Question 15:**String and Arithmetic Operations: Given x = "100" and y = "200", convert them to integers, add them, then convert the result back to a string. (5 Points)

**Question 17:** Given a = 5.7 and b = "3", convert b to an integer, add a and b, and round the result to the nearest integer. (5 Points)

**Question 19:** Develop a Python script that prompts the user to provide a speed in kilometers per hour and then converts it to meters per second (5 Points)
